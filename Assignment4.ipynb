# Visualization of employed persons in most populated states in Malaysia 

```python
%matplotlib notebook
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('bptms_Employed_by_state.csv') # the employed person dataset
df
```

```python
df_labour = pd.read_csv('bptms_Labour_force_by_state.csv') # the labour force dataset
df_labour
```

```python
df_2 = df.drop(df.index[187:561]) # get only the years 2006 to 2016
df_2.columns= ['State', 'Year', 'Employed(000)'] # rename the columns
df_2
```

```python
df_labour_2 = df_labour.drop(df_labour.index[187:561]) # get only the years 2006 to 2016
df_labour_2.columns= ['State', 'Year', 'Labour Force(000)'] # rename the columns 
df_labour_2
```

```python
states_to_keep = ['Selangor','Johor','Perak','Sarawak','Sabah'] # only keep the 5 most populated states

df_2 = df_2[df_2['State'].isin(states_to_keep)]
df_2.reset_index(drop='true',inplace =True)

df_labour_2 = df_labour_2[df_labour_2['State'].isin(states_to_keep)]
df_labour_2.reset_index(drop='true',inplace =True)
```

```python
#df_2['Employed(000)'] = df_2['Employed(000)'].str.replace(',', '')
df_2.loc[:, 'Employed(000)']=df_2['Employed(000)'].str.replace(',', '') # remove special characters in employed column
df_2
```

```python
df_labour_2.loc[:, 'Labour Force(000)']=df_labour_2['Labour Force(000)'].str.replace(',', '') # remove special characters in employed column
df_labour_2
```

```python
df_2.loc[:,'Employed(000)']=df_2['Employed(000)'].astype(float)
df_2.loc[:,'Year']=df_2['Year'].astype(int)

df_labour_2.loc[:,'Labour Force(000)']=df_labour_2['Labour Force(000)'].astype(float)
df_labour_2.loc[:,'Year']=df_2['Year'].astype(int)
```

```python
df_pivot= df_2.pivot_table(values='Employed(000)',index=['State'],columns='Year')
df_pivot
```

```python
labour_sum=df_labour_2.groupby('Year').sum()
df_labour_2 = df_labour_2.pivot_table(values=labour_sum,columns='Year')
df_labour_2
```

```python
df_pivot = df_pivot.T
df_labour_pivot = df_labour_2.T
ax1 = df_pivot.plot(colormap='jet', figsize=(12,5), lw=2);
ax2 = df_labour_pivot.plot(ax=ax1)
plt.xlabel('Years')
plt.ylabel('Number of people in thousands')
plt.title('Employed persons in popular states in Malaysia 2006 - 2016')
plt.show()
```
